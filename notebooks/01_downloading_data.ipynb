<h2>Download Tesla Stock Price Data<h2>

Download Tesla's stock price data from Yahoo Finance from 2022/02/01 to 2025/02/01

```python
import pandas as pd
import yfinance as yf

# define stock ticker and time period
ticker = "TSLA"
start_date = "2022-02-01"
end_date = "2025-02-28"

# download stock price data from Yahoo Finance
tesla_stock_data_raw = yf.download(ticker, start=start_date, end=end_date)

# reset index to get data as columns
tesla_stock_data_raw.reset_index(inplace=True)

# save as CSV
tesla_stock_data_raw.to_csv('/content/drive/MyDrive/Colab Notebooks/final_project/tesla_stock_data.csv', index=False)

# save as JSON
tesla_stock_data_raw.to_json('/content/drive/MyDrive/Colab Notebooks/final_project/tesla_stock_data.json', orient='records', date_format="iso")

print('Tesla stock price data saved successfuly')
```

<h2>Fetching News by G News API<h2>

Due to the limitation of the number of requests per day, I had to divide the process into four steps to fetch news over 3 past years.

```python
import requests
import json
import datetime
import time
import os

# G_News API key
API_KEY = "....."

# search query (keywords to filter news)
QUERY = """(Tesla AND carbon emissions) OR (Tesla AND sustainability) OR (Tesla AND green finance) OR
        (Tesla AND renewable energy) OR (Tesla AND battery technology) OR (Tesla AND clean energy) OR
        (Tesla AND ESG)"""


# G_News API base URL
BASE_URL = "https://gnews.io/api/v4/search"

# directory to save the data
SAVE_DIR = "...."
os.makedirs(SAVE_DIR, exist_ok=True)
```

```python
# Feb 2022 to Oct 2022

START_DATE = datetime.datetime(2022, 2, 1) # Feb 2022
END_DATE = datetime.datetime(2022, 10, 31) # Oct 2022
current_date = START_DATE

while current_date <= END_DATE:
    next_month = (current_date.replace(day=28) + datetime.timedelta(days=4)).replace(day=1)
    articles_data = []
    zero_article_days = 0

    day_iter = current_date
    while day_iter < next_month:
        from_date = day_iter.strftime("%Y-%m-%dT00:00:00Z")
        to_date = day_iter.strftime("%Y-%m-%dT23:59:59Z")

        params = {
            "q": QUERY,
            "in": "title,description,content",
            "from": from_date,
            "to": to_date,
            "max": 50,
            "expand": "content",
            "apikey": API_KEY
        }

        response = requests.get(BASE_URL, params=params)
        data = response.json()

        if response.status_code == 200 and "articles" in data and data["articles"]:
            for article in data["articles"]:
                articles_data.append({
                    "date": article["publishedAt"][:10],
                    "title": article["title"],
                    "description": article["description"],
                    "content": article.get("content", "Full content not available"),
                    "url": article["url"],
                    "source": article["source"]["name"],
                    "publishedAt": article["publishedAt"]
                })
        else:
            zero_article_days += 1

        day_iter += datetime.timedelta(days=1)
        time.sleep(0.2)  # API rate limit (max 6 requests/sec)

    filename = os.path.join(SAVE_DIR, f"tesla_news_{current_date.strftime('%Y-%m')}.json")
    with open(filename, "w", encoding="utf-8") as f:
        json.dump(articles_data, f, ensure_ascii=False, indent=4)

    print(f"{len(articles_data)} articles saved for {current_date.strftime('%Y-%m')}.")
    print(f"{zero_article_days} days had 0 articles.")

    current_date = next_month
```
